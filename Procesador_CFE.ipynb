```python
import pandas as pd
import numpy as np
import unicodedata

# Read the excel CFE and Proveedores files from GitHub
# Get the raw file URL by replacing 'blob' with 'raw' in the GitHub URL

url = "https://github.com/Fercaspe/CFE_Processor/raw/main/Facturas_Fuxion_DGI.xlsx"
url_p = "https://github.com/Fercaspe/CFE_Processor/raw/main/Cuentas_Fuxion.xlsx"

# Create dataframe ignoring first 8 rows with no relevant information & explicitly specify the engine (CFE is .xls)

df = pd.read_excel(url, skiprows=8, engine='openpyxl')

df_p = pd.read_excel(url_p)

df
```

```python
# Duplicate the fist row to use it as columns names
df.columns = df.iloc[0]

# Drop the row and reset the index
df = df[1:].reset_index(drop=True)

df
```

```python
# NORMALIZE COLUMNS NAMES

# Function to normalize the names

def normalize_names(text):
    text = str(text) # Convert to string
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')  # Special characters control
    text = text.replace(' ', '_')
    return text

# Call the function
df.columns = [normalize_names(col) for col in df.columns]

df
```

```python
df_p
```

```python
# Drpo duplicates in df_p
df_p = df_p.drop_duplicates(subset=['RUT_Emisor'])
```

```python
# Convert 'RUT_Emisor' column in both DataFrames to the same data type (string)

df['RUT_Emisor'] = df['RUT_Emisor'].astype(str)
df_p['RUT_Emisor'] = df_p['RUT_Emisor'].astype(str)
```

```python
df_merged = df.merge(df_p, on='RUT_Emisor', how='inner', indicator=True)
df_merged
```

```python
df_temp = df[["Numero", "Monto_Neto", "IVA_Ventas", "Monto_Total"]]
df_temp
```

```python
df_temp.rename(columns={'Monto_Neto':'1_Monto_Neto','IVA_Ventas':'2_IVA_Ventas','Monto_Total':'3_Monto_Total'},inplace=True)

# Transfor df_temp using melt

df_temp = df_temp.melt(id_vars=['Numero'], value_vars=['1_Monto_Neto', '2_IVA_Ventas', '3_Monto_Total'], var_name='Tipo_Monto', value_name='Monto')

# Sort by RUT and Tipo Monto

df_temp = df_temp.sort_values(by=['Numero', 'Tipo_Monto']).reset_index(drop=True)

df_temp = df_temp[['Numero', 'Monto', 'Tipo_Monto']]

# This code takes each value from the columns `Monto Neto`, `IVA Ventas` and `Monto Total` and converts them into individual rows under the `Monto` column, keeping the `RUT` as id.

df_temp
```

```python
# sort the entire df by 'Numero', and then triple each row so it will be equals the df_temp

df_main = df_merged.sort_values(by=['Numero']).reset_index(drop=True)

df_main = df_main.loc[df.index.repeat(3)].reset_index(drop=True)

df_main
```

```python
df_main = pd.concat([df_main, df_temp], axis=1)

df_main
```

```python
df_main.columns
```

```python
df_main = df_main.drop(columns=['Numero'])

df_main["NUMERO"] = df_temp["Numero"]

df_main.columns
```

```python
# Create a new df to receive all the data

total_filas = df_main.shape[0]

df_nuevo = pd.DataFrame(index=range(total_filas), columns=["DIA", "DEBE", "HABER", "CENTROCOSTOS", "RUT", "CONCEPTO", "MONEDA", "TOTAL", "CODIGODEIVA", "IVA", "LIBRO", "TIPOMONTO", "TIPOMONEDA", 'Cta_Proveedor_UYU', 'Cta_Proveedor_USD', 'Cta_Gasto_UYU', 'Cta_Gasto_USD', 'IVA_Ventas_UYU', 'IVA_Ventas_USD'])

df_nuevo.head()
```

```python
df_nuevo["DIA"] = df_main["Fecha_comprobante"]
df_nuevo["CENTROCOSTOS"] = df_main["Centro_Costos"]
df_nuevo["RUT"] = df_main["RUT_Emisor"]
df_nuevo["CONCEPTO"] = df_main["Tipo_CFE"] + " " + df_main["Serie"] + " " + df_main["NUMERO"] + " " + df_main["Proveedor"]
df_nuevo["TIPOMONEDA"] = df_main["Moneda"]
df_nuevo["TIPOMONTO"] = df_main["Tipo_Monto"]
df_nuevo["TOTAL"] = df_main["Monto"]
df_nuevo["CODIGODEIVA"] = df_main["Codigo_IVA"]
df_nuevo["IVA"] = df_main["IVA"]
df_nuevo["LIBRO"] = df_main["Libro"]
df_nuevo["Cta_Proveedor_UYU"] = df_main["Cta_Proveedor_UYU"]
df_nuevo["Cta_Proveedor_USD"] = df_main["Cta_Proveedor_USD"]
df_nuevo["Cta_Gasto_UYU"] = df_main["Cta_Gasto_UYU"]
df_nuevo["Cta_Gasto_USD"] = df_main["Cta_Gasto_USD"]
df_nuevo["IVA_Ventas_UYU"] = df_main["IVA_Ventas_UYU"]
df_nuevo["IVA_Ventas_USD"] = df_main["IVA_Ventas_USD"]

df_nuevo.head(9)
```

```python
df_nuevo['MONEDA'] = np.where(df_nuevo['TIPOMONEDA'] == 'UYU', 0, 1)
df_nuevo.head()
```

```python
def determinar_cuenta_contable_debe(row):
    if row['MONEDA'] == 0 and row['TIPOMONTO'] == '1_Monto_Neto':
        return row['Cta_Gasto_UYU']
    elif row['MONEDA'] == 0 and row['TIPOMONTO'] == '2_IVA_Ventas':
        return row['IVA_Ventas_UYU']
    elif row['MONEDA'] == 1 and row['TIPOMONTO'] == '1_Monto_Neto':
        return row['Cta_Gasto_USD']
    elif row['MONEDA'] == 1 and row['TIPOMONTO'] == '2_IVA_Ventas':
        return row['IVA_Ventas_USD']
    else:
         return ""


def determinar_cuenta_contable_haber(row):
  if row['MONEDA'] == 0 and row['TIPOMONTO'] == '3_Monto_Total':
    return row['Cta_Proveedor_UYU']
  elif row['MONEDA'] == 1 and row['TIPOMONTO'] == '3_Monto_Total':
    return row['Cta_Proveedor_USD']
  else:
    return ""


df_nuevo['DEBE'] = df_nuevo.apply(determinar_cuenta_contable_debe, axis=1)
df_nuevo['HABER'] = df_nuevo.apply(determinar_cuenta_contable_haber, axis=1)

df_nuevo.head(15)
```

```python
df_salida = df_nuevo[["DIA", "DEBE", "HABER", "CENTROCOSTOS", "RUT", "CONCEPTO", "MONEDA", "TOTAL", "CODIGODEIVA", "IVA", "LIBRO"]]
df_salida.head(10)
```
