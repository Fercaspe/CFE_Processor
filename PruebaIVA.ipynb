```python
import pandas as pd
import numpy as np

# Leer archivo excel desde GitHub
# Get the raw file URL by replacing 'blob' with 'raw' in the GitHub URL

url = "https://github.com/Fercaspe/CFE_Processor/raw/main/cferecibidos.xls"

# Creo el Dataframe saltando las primeras 8 filas que no tienen datos relevantes & Explicitly specify the engine

df = pd.read_excel(url, skiprows=8, engine='xlrd')

df
```

```python
# Duplico la primera fila cómo nombre de columna
df.columns = df.iloc[0]

# Borro la fila usada y reinicio el index
df = df[1:].reset_index(drop=True)

df
```
