```python
import pandas as pd
import numpy as np
import unicodedata

# Read the excel CFE and Proveedores files from GitHub
# Get the raw file URL by replacing 'blob' with 'raw' in the GitHub URL

url = "https://github.com/Fercaspe/CFE_Processor/raw/main/cferecibidos.xls"
url_p = "https://github.com/Fercaspe/CFE_Processor/raw/main/proveedores.xlsx"

# Create dataframe ignoring first 8 rows with no relevant information & explicitly specify the engine (CFE is .xls)

df = pd.read_excel(url, skiprows=8, engine='xlrd')

df_p = pd.read_excel(url_p)

df.head()
```

```python
# Duplicate the fist row to use it as columns names
df.columns = df.iloc[0]

# Drop the row and reset the index
df = df[1:].reset_index(drop=True)

df.head()
```

```python
# NORMALIZE COLUMNS NAMES

# Function to normalize the names

def normalize_names(text):
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')  # Special characters control
    text = text.replace(' ', '_')
    return text

# Call the function
df.columns = [normalize_names(col) for col in df.columns]

df.head()

```

```python
df_p.head()
```

```python
# Convert 'RUT_Emisor' column in both DataFrames to the same data type (string)
df['RUT_Emisor'] = df['RUT_Emisor'].astype(str)
df_p['RUT_Emisor'] = df_p['RUT_Emisor'].astype(str)

# Now perform the merge
df_main = pd.merge(df, df_p, how='left', on='RUT_Emisor')

df_main.head()
```

```python
# repeat each row 3 times to unify structure to output file

df_repetido = df_main.loc[df.index.repeat(3)].reset_index(drop=True)

df_repetido.head()
```
